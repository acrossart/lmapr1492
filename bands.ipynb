```python
from pymatgen.ext.matproj import *  
from pymatgen.electronic_structure.bandstructure import *  
from pymatgen.electronic_structure.plotter import BSPlotter
from pymatgen.io.cif import CifParser
with MPRester("rjzlkOjkFDW0uIctnb2o") as m:
    bandstructure =m.get_bandstructure_by_material_id("mp-1070")

#Determination de la bande interdite    
band_gap=bandstructure.get_band_gap()

print('information sur la bande interdite:',band_gap)
#Plotter le grphe
bsplotter = BSPlotter(bandstructure)
bsplotter.show()



```

```python
#Calcul de la masse effective
#h bar en eV
h_bar=6.58*10**(-16)
```

```python
#Pour le minimum de la premiere bande
cbm=bandstructure.get_cbm()

E_cbm = cbm['energy']
#Vecteur de base
a=4.394

G = 2*np.pi/a
#formule utilise pour calculer la masse effective
m_effective = h_bar**2*G**2/(2*E_cbm)
print('La masse effective pour le minimum de la premiere bande (en kg):',m_effective)
```

```python
#Pour le maximum de la derniere bande
vbm=bandstructure.get_vbm()

E_vbm = vbm['energy']
#Vecteur de base
a=4.394

G = 2*np.pi/a
#formule utilise pour calculer la masse effective
m_effective = h_bar**2*G**2/(2*E_vbm)
print('La masse effective pour le maximum de la derniere bande ( en kg):',m_effective)
```
